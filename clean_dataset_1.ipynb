```python
import pandas as pd
import numpy as np
import math
import matplotlib as plt
import pdpipe as pdp
```

```python
name = "dataset_1"

original_df = pd.read_csv(name+".csv")
df = original_df
df.head()
```

```python
pipe = pdp.ColDrop("bikeid")
pip += pdp.DropNa(subset=["start_station_id","end_station_id","tripduration"])
pipeline += pdp.RowDrop({"tripduration": lambda t : t<60})

df = pipeline(df)
df.head()
```
